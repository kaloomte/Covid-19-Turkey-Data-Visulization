```python
#Read dataset
egitimdf = pd.read_csv('Covid-19/pivot.csv', delimiter='|', header=None)
```

```python
#Clean the dataset from which has NaN 
#Fill with '0'
egitimdf.drop([0,1,2,3,82], inplace=True)
egitimdf.drop([0,16,2], axis=1, inplace=True)
egitimdf.fillna(0, inplace=True)
```

```python
#Created new Column names list
col_list = ['Durum', 'Akdeniz','Batı Anadolu',
 'Batı Karadeniz','Batı Marmara','Doğu Karadeniz',
 'Doğu Marmara','Ege','Güneydoğu Anadolu','Kuzeydoğu Anadolu','Orta Anadolu','Ortadoğu Anadolu','Türkiye','İstanbul']
#Set the column names
egitimdf.columns = col_list
```

Let's take a quick look at what the data looks like

```python
#All columns except 'Durum' converted into numeric
egitimdf.loc[:, ~egitimdf.columns.isin(['Durum'])] = egitimdf.loc[:, ~egitimdf.columns.isin(['Durum'])].apply(pd.to_numeric).astype(int)
```

```python
bilmeyen = pd.Series(egitimdf[egitimdf['Durum'].str.contains("Bilmeyen")].sum())
bilen = pd.Series(egitimdf[egitimdf['Durum'].str.contains("Bilen")].sum())
bilinmeyen = pd.Series(egitimdf[egitimdf['Durum'].str.contains("Bilinmeyen")].sum())

kadın_bilen = pd.Series(egitimdf[egitimdf['Durum'].str.contains("Kadın") & egitimdf['Durum'].str.contains("Bilen")].sum())
kadın_bilmeyen = pd.Series(egitimdf[egitimdf['Durum'].str.contains("Kadın") & egitimdf['Durum'].str.contains("Bilmeyen")].sum())
kadın_bilinmeyen = pd.Series(egitimdf[egitimdf['Durum'].str.contains("Kadın") & egitimdf['Durum'].str.contains("Bilinmeyen")].sum())

erkek_bilen = pd.Series(egitimdf[egitimdf['Durum'].str.contains("Erkek") & egitimdf['Durum'].str.contains("Bilen")].sum())
erkek_bilmeyen = pd.Series(egitimdf[egitimdf['Durum'].str.contains("Erkek") & egitimdf['Durum'].str.contains("Bilmeyen")].sum())
erkek_bilinmeyen = pd.Series(egitimdf[egitimdf['Durum'].str.contains("Erkek") & egitimdf['Durum'].str.contains("Bilinmeyen")].sum())
```

As it seems there are informations about pandemic until 16 July in "update" csv file. Required really update until 20 August. 

```python
new_egitim_df = pd.DataFrame([bilen,bilmeyen, bilinmeyen, kadın_bilen, kadın_bilinmeyen, kadın_bilmeyen, erkek_bilen,erkek_bilinmeyen,erkek_bilmeyen])
```

```python
index_list = ["bilen","bilmeyen", "bilinmeyen", "kadın_bilen", "kadın_bilinmeyen", "kadın_bilmeyen", "erkek_bilen","erkek_bilinmeyen","erkek_bilmeyen"]
```

```python
new_egitim_df.index = index_list
```

```python
new_egitim_df.drop('Durum',axis=1, inplace=True)
```
